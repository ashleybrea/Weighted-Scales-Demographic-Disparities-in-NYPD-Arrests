```python
# reading NY ARRESTS
nyar_df = pd.read_csv('/content/NYPD_Arrests_Data_ML.csv',on_bad_lines="warn",engine='python')

nyar_df.info()
```

```python
# selecting cols
nyar_df = nyar_df[['ARREST_DATE', 'LAW_CODE','ARREST_BORO','ARREST_PRECINCT','AGE_GROUP','PERP_SEX','PERP_RACE']]

# cleaning empty rows
nyar_df = nyar_df.dropna()

nyar_df.info()
```

```python
# updating ARREST_DATE into datetime
nyar_df['ARREST_DATE'] = pd.to_datetime(nyar_df['ARREST_DATE'])

nyar_df.head(5)
```

```python
# updating 'object' to 'category'
for col in ['LAW_CODE','ARREST_BORO','ARREST_PRECINCT','AGE_GROUP','PERP_SEX','PERP_RACE']:
    nyar_df[col] = nyar_df[col].astype('category')

nyar_df.info()
```

```python
# filtering arrests of minors from database
nyar_df = nyar_df[nyar_df['AGE_GROUP'] != '<18']

nyar_df.info()
```
